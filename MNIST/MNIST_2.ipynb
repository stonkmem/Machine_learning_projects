Here is my code for running the Fashion_MNIST dataset. Sure hope it's gonna be fun :)

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
import numpy
print("Done!")
```

Now, let's load Fashion MNIST.

```python
dataset = keras.datasets.mnist
(f_train, l_train), (f_test, l_test) = dataset.load_data()
f_train = f_train/255
f_test = f_test/255
print("Processed Data!")
```

Now that we've got that done, let's move onto coding our neural network.

```python
model = keras.models.Sequential([
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
print("Model made!")
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=["accuracy", "sparse_categorical_crossentropy"])
print("Compiled!")
```

#**DRAMATIC MUSIC PLAYS**

```python
model.fit(x=f_train, y=l_train, epochs=500)
t_loss, t_accuracy, t_sparse_categorical_crossentropy = model.evaluate(x=f_test, y=l_test)
print(f"Test loss = {t_loss}")
print(f"Test accuracy = {t_accuracy}")
print(f"Test sparse categorical crossentropy = {t_sparse_categorical_crossentropy}")
```
